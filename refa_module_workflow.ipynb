The `refa` module offers necessary classes to build a complete techno-economic analysis of different capacity upgrade options.

```python
from refa import (
    LineDesign, Conductor, Line, Rebuild, Reconductoring, VoltageUpgrade, HVDC, Existing, Analysis
)
```

Default constuctors are included in `refa.defaults` to illustrate how different inputs can be entered.

```python
from refa.defaults import (
    default_economics,
    default_clear_environment,
    default_conductor,
    default_structure_config_ac,
    default_structure_config_dc
)

my_conductor = default_conductor()
my_environment = default_clear_environment()
my_structure_config_ac = default_structure_config_ac()
my_structure_config_dc = default_structure_config_dc()
my_economics = default_economics()
```

Standard NESC 250B rule loading profiles are provided at `refa.standards`.

```python
from refa.standards import nesc_250b_heavy

my_loading = nesc_250b_heavy()
print(my_loading)
```

To create a `Line`, you should build a `LineDesign` and a `Conductor`. Typical conductor sizes can be found in `refa.defaults`.

```python
from refa.defaults import acsr_795_0_drake, acss_795_0_cuckoo, accc_1035_dublin, load_conductors_from_csv

line_design1 = LineDesign(
    environment=my_environment,
    nbr_circuits=1,
    nbr_bundles=3,
    nbr_conds_per_bundle=1,
    length_km=10,
    avg_span_m=250,
    max_span_m=300
)
line_design2 = LineDesign(
    environment=my_environment,
    nbr_circuits=1,
    nbr_bundles=3,
    nbr_conds_per_bundle=1,
    length_km=30,
    avg_span_m=300,
    max_span_m=300
)
print("Line design -> ", line_design1)

conductor = Conductor(
        type='ACSR',
        code='266.8_PARTRIDGE',
        area_mm2=157.0,
        diameter_mm=16.0,
        weight_n_per_m=5.36,
        conductor_rts_kn=50.26,
        dol_per_1000_ft=735.0,
        inst_dol_per_1000_ft=1027.0,
        accessories_dol_per_1000_ft=263.0,
        str_costs_dol=0.0,
        temp_dc_c=20.0,
        temp_low_c=25.0,
        temp_high_c=75.0,
        max_temperature_c=100.0,
        res_dc_ohm_per_m=0.000209,
        res_low_ohm_per_m=0.000209,
        res_high_ohm_per_m=0.000256,
        elastic_modulus_gpa=75.5,
        coeff_thermal_expan_per_cel=1.92e-05,
        emissivity=0.5,
        solar_absorptivity=0.5
)
print("Conductor using constructor -> ", conductor)

conductor1 = acss_795_0_cuckoo()
conductor2 = acsr_795_0_drake()
conductor3 = accc_1035_dublin()
print("Conductor from default database -> ", conductor1)

conductors = load_conductors_from_csv('conductors.csv')
print("Conductor from csv file -> ", conductors.acsr_477_0_pelican())
```

Then, a line can be created either using `Line` constructor, or directly summing a `LineDesign` and a `Conductor`.

```python
line1 = Line(line_design=line_design1, conductor=conductor1)
line2 = line_design2 + conductor2
line3 = conductor3 + line_design1

print(line1)
```

Different technical calculations can be conducted using a `Line` object.

```python
temperature_from_current = line1.temperature_at_current(current_a=1400) 
resistance_from_current = line1.resistance_at_current(current_a=1400)
temperature_from_p_and_v = line1.temperature_at_current(power_mw=550, voltage_kv=345)
resistance_from_p_and_v = line1.resistance_at_current(power_mw=550, voltage_kv=345)
print(f"Conductor temperature when current is provided -> {temperature_from_current} deg. Cel")
print(f"Conductor resistance when current is provided -> {resistance_from_current} Ohm/m")
print(f"Conductor temperature when power and voltage are provided -> {temperature_from_p_and_v} deg. Cel")
print(f"Conductor resistance when power and voltage are provided -> {resistance_from_p_and_v} Ohm/m")

ampacity = line1.ampacity_at_environment()
print(f"Ampacity -> {ampacity} Amps")

sag_from_peak_current = line1.sag_at_current(current_a=1400, initial_tension_percentage=0.2)
sag_from_power_and_voltage = line1.sag_at_power_and_voltage(power_mw=500, voltage_kv=230, initial_tension_percentage=0.2)
sag_from_conductor_temperature = line1.sag_at_temperature(temp_at_current_c=85, initial_tension_percentage=0.2)
sag_from_loading_profile = line1.sag_at_loading(loading_conditions=my_loading, initial_tension_percentage=0.2)
print(f"Sag when peak current is provided -> {sag_from_peak_current}")
print(f"Sag when conductor temperature is provided -> {sag_from_conductor_temperature}")
print(f"Sag when a wind/ice loading profile is provided -> {sag_from_loading_profile}")

corona_inception_voltage = line1.corona_inception_voltage(structure_config=my_structure_config_ac) 
corona_voltage_gradient = line1.corona_voltage_gradient(structure_config=my_structure_config_ac)
print(f"Corona Inception voltage -> {corona_inception_voltage} kV")
print(f"Corona voltage gradient -> {corona_voltage_gradient} kV/cm")
```

Also, losses and congestion of a given `Line` can be calculated using available functions.

```python
resistive_line_losses_from_current = line1.resistive_line_losses(current_a=1400, load_factor=0.6)
resistive_line_losses_with_losses = \
    line1.resistive_line_losses_considering_congestion(power_mw=550, voltage_kv=230, load_factor=0.6)
print(f"Resistive line losses when current is provided -> {resistive_line_losses_from_current} MWh/m")
print(f"Resistive line losses when power is provided -> {resistive_line_losses_with_losses} MWh/m")

corona_discharge_losses = line1.corona_discharge_losses(
    voltage_kv=345, load_factor=0.6, structure_config=my_structure_config_ac
)
print(f"Corona discharge losses -> {corona_discharge_losses} MWh/m")

corona_discharge_losses_dc = line1.corona_discharge_losses(
    voltage_kv=345, load_factor=0.6, structure_config=my_structure_config_dc, is_hvdc=True
)
print(f"Corona discharge losses (DC line) -> {corona_discharge_losses_dc} MWh/m")

congestion_from_current = line1.congestion(current_a=1500, voltage_kv=230)
congestion_from_power = line1.congestion(power_mw=700, voltage_kv=230)
print(f"Congestion when current is provided -> {congestion_from_current} MW")
print(f"Congestion when power is provided -> {congestion_from_power} MW")
```

The overall technical performance of a `Line` can be calculated in a single function call. The basic usage calculates only ampacity and sag when current or (power+voltage) are provided; then includes losses if load_factor is provided, and corona discharge if structure_config is provided.

```python
technical_perf_from_current = line1.overall_technical_performance(current_a=1500)
print(f"Overall performance when only current is provided -> {technical_perf_from_current}")

technical_perf_from_power_and_voltage = line1.overall_technical_performance(
    power_mw=500, voltage_kv=230
)
print(f"Overall performance when power and voltage are provided -> {technical_perf_from_power_and_voltage}")

technical_perf_including_losses = line1.overall_technical_performance(
    current_a=1500, load_factor=0.6
)
print(f"Overall performance including losses -> {technical_perf_including_losses}")

technical_perf_including_losses_and_corona = line1.overall_technical_performance(
    current_a=1500, voltage_kv=230, load_factor=0.6, structure_config=my_structure_config_ac
)
print(f"Overall performance including corona discharge -> {technical_perf_including_losses_and_corona}")
```

Different types of projects can be created using `a LineDesign + a list of Conductor objects` or `a list of Line Objects`.

```python
rebuild_prj = Rebuild(
    conductor_list=[acsr_795_0_drake(), acss_795_0_cuckoo(), accc_1035_dublin()], 
    line_design=line_design2,
    economics=my_economics,
    power_mw=400,
    voltage_kv=230,
    structure_config=my_structure_config_ac
)
reconductoring_prj = Reconductoring(
    line_list=[line1, line2], 
    economics=my_economics,
    power_mw=400,
    voltage_kv=230,
    structure_remaining_life=25,
    structure_config=my_structure_config_ac
)
voltageupgrade_prj = VoltageUpgrade(
    line_list=[line2, line1], 
    economics=my_economics,
    power_mw=400,
    voltage_kv=345, 
    structure_remaining_life=0, 
    conductor_remaining_life=0,
    cost_substations_upgrade_dol=2000000,
    structure_config=my_structure_config_ac  
)
hvdc_prj = HVDC(
    line_list=[line3, line2], 
    economics=my_economics,
    power_mw=400,
    voltage_kv=500, 
    structure_remaining_life=0, 
    conductor_remaining_life=0, 
    cost_converters_dol=3000000,
    nbr_dc_poles_per_circuit=2,
    structure_config=my_structure_config_dc
)
existing_project = Existing(
    line_list=[line1], 
    economics=my_economics,
    power_mw=400,
    voltage_kv=230,
    structure_remaining_life=25,
    conductor_remaining_life=15,
    structure_config=my_structure_config_ac
)
```

VoltageUpgrade and HVDC projects contain costs that are not present in other types of projects. These costs of substation, transformer, or converter modifications are required by VoltageUpgrade and HVDC object constructors. These costs are currently added to the upfront capital cost, and can be reset as follows.

```python
voltageupgrade_prj.cost_substations_upgrade_dol = 1000000
hvdc_prj.cost_converters_dol = 2000000
```

Multiple projects can be compared using an object `Analysis` based on total net-present value of cost, or marginal costs of structures, conductors, losses, and congestion. 
By default, each line in a project is checked for feasibility in terms of ampacity (if `select_ampacity_feasible=True`), sag (if `select_sag_feasible=True`), and corona (if `select_corona_feasible=True`). This feasibility check can be deactivated by setting these attributes of project objects to `False`.

```python
refa = Analysis(project_list=[
    rebuild_prj, reconductoring_prj, voltageupgrade_prj, hvdc_prj, existing_project
    ])
refa.total_costs_of_projects(time_horizon=70)
# refa.total_costs_of_projects_including_losses(time_horizon=70, load_factor=0.6)

# refa.structure_costs_of_projects(time_horizon=70)
# refa.conductor_costs_of_projects(time_horizon=70, select_feasible=False)
# refa.losses_costs_of_projects(time_horizon=70, load_factor=0.6)
# refa.congestion_costs_of_projects(time_horizon=70)
```
